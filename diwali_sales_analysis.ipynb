```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Diwali Sales Data.csv", encoding="unicode_escape")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
data_set=[['madhav' , 1], ['Anshul' , 3], ['Prakhar', 8]]
df_set = pd.DataFrame(data_set, columns=['Name' , 'Age'])
df_set
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_statuus' : 'Shadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax = sns.countplot(x='Gender' , data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
```

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_by_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='State', y='Orders', data= sales_by_state)
```

```python
sales_by_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='State', y='Amount', data= sales_by_state)
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize' :(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y='Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize' :(20,5)})
sns.barplot(data = sales_state, x = 'Occupation', y='Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize' :(30,10)})
sns.barplot(data = sales_state, x = 'Product_Category', y='Amount', hue='Gender')
```

```python
sales_state=df.groupby(['Product_ID' , 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize' :(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y='Orders')
```
